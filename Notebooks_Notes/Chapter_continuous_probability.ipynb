```python
dist = stats.norm(0, 1)#mean 0, standard deviation 1
a=-1.5
b=0.3
Prob=dist.cdf(b)-dist.cdf(a)
print("Probability computed via CDF",Prob)
print("")
samples=dist.rvs(30)#we generate 30 random numbers from dist
print("Probability computed via sampling",len(np.where((a<=samples) & (samples<=b))[0])/len(samples))
print("")
samples=dist.rvs(50)#we generate 50 random numbers from dist
print("Probability computed via sampling",len(np.where((a<=samples) & (samples<=b))[0])/len(samples))
print("")
samples=dist.rvs(500000)#we generate 500000 random numbers from dist
print("Probability computed via sampling",len(np.where((a<=samples) & (samples<=b))[0])/len(samples))
```

```python
dist = stats.norm(0, 1)#mean 0, standard deviation 1
samples=dist.rvs(500000)#we generate 500000 random numbers from dist
print("Mean",np.sum(samples)/len(samples))
mu=np.sum(samples)/len(samples)
print("Variance",np.sum((samples-mu)**2)/len(samples))
print("Standard Deviation",np.sqrt(np.sum((samples-mu)**2)/len(samples)))
dist = stats.norm(-2, 2)#mean 2, standard deviation 2
samples=dist.rvs(500000)#we generate 500000 random numbers from dist
print("Mean",np.sum(samples)/len(samples))
mu=np.sum(samples)/len(samples)
print("Variance",np.sum((samples-mu)**2)/len(samples))
print("Standard Deviation",np.sqrt(np.sum((samples-mu)**2)/len(samples)))
```

## Conditional probability via sampling

```python
import scipy as sp
import numpy as np
import arviz as az
def event(x,interval):
    return np.logical_and(np.greater(x,interval[0]),np.less(x,interval[1]))

interval=[0.25,np.inf]# event A, w>0.25

dist = stats.norm(0.25, 0.05)#mean 0.25, standard deviation 0.05
samples=dist.rvs(100000)#we generate 10000 random numbers from dist
ind=event(samples,interval)#select the samples that belong to the interval A
samples=samples[ind]
print("number of samples that are in the interval [0.25,$\infty$):",len(samples))
plt.hist(samples,50)#this is the posterior PDF
plt.xlabel("weight (Kg)",fontsize=16)
plt.ylabel("counts",fontsize=16)
plt.savefig("posterior_weightcake_hist.pdf")
plt.figure()
az.plot_dist(samples)#this is the posterior PDF
plt.xlabel("weight (Kg)",fontsize=16)
plt.ylabel("p(w|A)",fontsize=16)
print("P(B|A)",np.sum(event(samples,[0.2,0.3]))/len(samples))
plt.tight_layout()
plt.savefig("posterior_weightcake.pdf")
```

```python
import pymc3 as pm
import theano.tensor as tt
import arviz as az
def event_A(x,interval):
    return pm.math.log(pm.math.sigmoid((x-interval[0])*5000.0)*pm.math.sigmoid((-x+interval[1])*5000.0))

interval=[0.25,np.inf]# event A, w>0.25
mymodel=pm.Model()
with mymodel:
    x = pm.Normal('x',0.25,0.05)
    pm.DensityDist('event',event_A,observed=dict(x=x, interval=interval))
    
with mymodel:
    samples=pm.sample(30000,tune=10000, chains=1)
    
```

```python
print("number of samples that are in the interval [0.25,$\infty$):",len(samples))
plt.hist(samples['x'],50)#this is the posterior PDF
plt.xlabel("weight (Kg)",fontsize=16)
plt.ylabel("counts",fontsize=16)
plt.figure()
az.plot_dist(samples['x'])#this is the posterior PDF
plt.xlabel("weight (Kg)",fontsize=16)
plt.ylabel("p(w|A)",fontsize=16)
print("P(B|A)",np.sum((samples['x']>=0.2)&(samples['x']<= 0.3))/len(samples))
plt.tight_layout()
```

```python
mean = 0.25

standard_deviation = 0.05


x_values = np.linspace(6.5,9.5,100)

dist = stats.norm(8, 0.3)

plt.plot(x_values, dist.pdf(x_values))
plt.xlabel("length (cm)", fontsize=16);
plt.ylabel("p(x)", fontsize=16);
plt.savefig("marginal_x.pdf")
plt.figure()
plt.plot(x_values, dist.pdf(x_values))
plt.xlabel("width (cm)", fontsize=16);
plt.ylabel("p(y)", fontsize=16);
plt.savefig("marginal_y.pdf")
```

```python
# Create meshgrid
r=np.linspace(6.5,9.5)
xx, yy = np.meshgrid(r,r)
dist = stats.multivariate_normal([8,8],np.array([[0.25,-0.24],[-0.24,0.25]]))
vv = np.vstack([xx.ravel(), yy.ravel()])
f = np.reshape(dist.pdf(vv.T), xx.shape)

fig = plt.figure(figsize=(13, 7))
ax = plt.axes(projection='3d')
w = ax.plot_wireframe(xx, yy, f)
ax.set_xlabel('x',fontsize=16)
ax.set_ylabel('y',fontsize=16)
ax.set_zlabel('PDF',fontsize=16)
plt.savefig("fish2d0.pdf")
plt.figure()
cfset = plt.contourf(xx, yy, f)
plt.xlabel('x',fontsize=16)
plt.ylabel('y',fontsize=16)
plt.axis('square')
plt.savefig("fish2d_cont3n.pdf")
```
